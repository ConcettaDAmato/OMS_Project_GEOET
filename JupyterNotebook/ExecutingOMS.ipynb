```python
import os
import sys
import subprocess
```

```python
#To be added to geoframe.py
# Improvable with more controls and more pythonic setting of global variables
# Using https://stackabuse.com/executing-shell-commands-with-python/
def OMS_run_MAC(simulation):
    sim=simulation+".sim"
    if(JAVA_HOME==''):
        print("JAVA_HOME not set")
        return -1
    else:
        com=JAVA_HOME+'java'
    if(console_location ==''):
        print("console location not found")
        return -1
    proj_loc=os. getcwd()
    total_cmd=[com,'-Doms3.work='+proj_loc,'-cp','.:'+console_location+'/*:lib/*:dist/*','oms3.CLI', '-r','simulation/'+sim]
    useless_cat_call = subprocess.Popen(total_cmd, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    output, errors = useless_cat_call.communicate(input="Hello from the other side!")
    useless_cat_call.wait()
    return output,errors

def OMS_run_WINDOWS(simulation):
    sim=simulation+".sim"
    if(JAVA_HOME==''):
        print("JAVA_HOME not set")
        return -1
    else:
        com=JAVA_HOME+'java'
    if(console_location ==''):
        print("console location not found")
        return -1
    proj_loc=os. getcwd()
    total_cmd=[com,'-Doms3.work='+proj_loc,'-cp','.;'+console_location+'/*;lib/*;dist/*','oms3.CLI', '-r','simulation/'+sim]
    useless_cat_call = subprocess.Popen(total_cmd, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    output, errors = useless_cat_call.communicate(input="Hello from the other side!")
    useless_cat_call.wait()
    return output,errors
```

```python
JAVA_HOME="/Users/concettadamato/opt/anaconda3/envs/geoframe_vincennes/lib/jvm/bin/"
console_location = "/Users/concettadamato/OMS/consoleOMS/oms-3.6.28-console/"

proj_location = "/Users/concettadamato/OMS/OMS_Project_GEOET"
os.chdir(proj_location)
#Actual Execution
output,errors=OMS_run_MAC('04_ProsperoPM')
print(output)
```
